```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
df_diabetes=pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
df_diabetes.head()
```

```python
X=df_diabetes
y=dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=10)
```

```python
X_train.head()
```

```python
y
```

```python
#correlation
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,15))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
treeregressor=DecisionTreeRegressor()
treeregressor.fit(X_train,y_train)
```

## Hyperparamter Tuning

```python
params={
    'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,6],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(DecisionTreeRegressor(),params,refit=True,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.best_params_
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error, mean_squared_error
```

```python
print(r2_score(y_test,grid_pred))
print(mean_absolute_error(y_test,grid_pred))
print(mean_squared_error(y_test,grid_pred))
```

```python
selectedmodel=DecisionTreeRegressor(criterion='poisson',max_depth=2, max_features='sqrt',splitter='best')
```

```python
selectedmodel.fit(X_train,y_train)
```

```python
y_pred=selectedmodel.predict(X_test)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(10,15))
tree.plot_tree(selectedmodel, filled=True)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
